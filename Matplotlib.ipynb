```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [10,20,15,25]

plt.plot(x,y)
plt.show()
```

```python
x = ["Mon","Tues","wed","Thru","Frid"]
y = [10,7,13,20,8]
plt.plot(x,y)

plt.title("Bakery Sales This Week!!")

plt.xlabel("Day")
plt.ylabel("Sales Per Day")

plt.show()
```

```python
import matplotlib.pyplot as plt
months = [1,2,3,4]
sales =[100,400,300,700]

plt.plot(months,sales,color="Red",linestyle="--",
         linewidth = 1 ,marker= "+",label ="2025 Sales Data")
plt.title("Sales Data")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend(loc ="upper left")
plt.grid(linestyle=":")
plt.xlim(1,4)
plt.xticks([1,2,3,4],["M1","M2","M3","M4"])
plt.show()
```

Bar Chart

```python
product = ["A","B","C","D"]
Sales = [1000,1500,1200,800]

plt.barh(product,Sales,color = 'orange',label ='sales 2025')
plt.show()
```

```python
region = ["East","West","North","South"]
sales = [466,232,577,180]

plt.pie(sales,labels=region,colors=['red','green','blue','grey'],autopct= '%1.1f%%')
plt.show()
```

```python
scores =[34,34,66,88,35,89,45,88,52,79,54,23]
plt.hist(scores,bins=7,color='orange',edgecolor ='black')
plt.show()
```

```python
hours = [1,2,3,4]
scores = [34,56,68,71]
plt.scatter(hours,scores,color='blue',marker='o',label ="A")
region = [1,2,3,4]
sales = [12,23,45,56]
plt.scatter(region,sales,color='green',marker='s',label="B")
plt.legend()
plt.grid()
plt.show()
```

```python
x = [1,2,3,4]
y = [10,20,30,40]

plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("Line Chart")

plt.subplot(1,2,2)
plt.bar(x,y)
plt.title("Bar Chart")

plt.show()
```

```python
fig,ax = plt.subplots(1,2,figsize=(10,5))

x = [1,2,3,4]
y = [10,20,30,40]

ax[0].plot(x,y)
ax[0].set_title("Line plot")

ax[1].bar(x,y)
ax[1].set_title("Bar Chart")

plt.show()
```

```python
x =[1,2,3,4]
y =[12,35,23,76]

plt.plot(x,y,color='blue',marker='o')
plt.title("Simple line Plot")
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.savefig('line_plt.png',dpi = 300,bbox_inches='tight')
plt.show()
```
